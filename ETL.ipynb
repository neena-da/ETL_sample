```python
# Import dependencies
%config Completer.use_jedi = False

import pandas as pd
from sqlalchemy import create_engine
from config import *
```

```python
# Read the temperature file
temp_csv_file = "Environment_Temperature_change.csv"
temp_df = pd.read_csv(temp_csv_file, encoding='iso-8859-1')
temp_df.head()
```

```python
# Dropping columns Y1961 to Y1989
temp_df.drop(temp_df.columns[7:36], axis=1, inplace=True)
```

```python
# Dropping columns
temp_df.drop(['Months Code', 'Element Code', 'Y2015', 'Y2016', 'Y2017', 'Y2018', 'Y2019'], axis = 1, inplace=True)
```

```python
temp_df.head()
```

```python
# Filtering dataframe to get the necessary rows
temp_filter_df = temp_df.loc[(temp_df['Area Code'] < 5000) & (temp_df['Months'] == 'Meteorological year') & 
                          (temp_df['Element'] == 'Temperature change'), :]
temp_filter_df
```

```python
# Dropping columns
temp_filter_df.drop(['Area Code', 'Months'], axis = 1, inplace=True)
temp_filter_df
```

```python
# Flipping the Year columns to rows
temp_flip_df = temp_filter_df.melt(id_vars=["Area", "Element", "Unit"], var_name="Year", value_name="Temp")
temp_flip_df
```

```python
# Stripping the 'Y' from the year column
temp_flip_df['Year'] = temp_flip_df['Year'].str.strip('Y')
temp_flip_df
```

```python
# Dropping na from the temperature datframe
clean_temp_df = temp_flip_df.dropna()
clean_temp_df
```

```python
# Renaming temperature dataframe columns
final_temp_df = clean_temp_df.rename(columns={'Area': 'country', 'Element': 'element', 'Unit': 'unit',
                                             'Year': 'year', 'Temp': 'temp'})
final_temp_df
```

```python
final_temp_df_new = final_temp_df.reset_index(drop=True)
final_temp_df_new
```

```python
final_temp_df_new1 = final_temp_df_new.reset_index()
final_temp_df_new1
```

```python
# Setting index column
final_temp_df_new1 = final_temp_df_new1.rename(columns={'index': 'id'})
final_temp_df_new1 = final_temp_df_new1 .set_index('id')
final_temp_df_new1
```

```python
# Reading Emissions csv file
emission_file = "greenhouse_gas_inventory.csv"
emission_df = pd.read_csv(emission_file)
emission_df.head()
```

```python
# Renaming and filtering columns
emission_filter_df = emission_df[['country_or_area', 'year', 'value']]
emission_filter_df = emission_filter_df.rename(columns={'country_or_area':'country', 'year': 'year', 'value': 'co2_emissions'})
emission_filter_df['co2_emissions'] = emission_filter_df['co2_emissions'].round(3)
emission_filter_df
```

```python
# Setting index
emission_filter_df['year'] = emission_filter_df['year'].astype(str)

emission_filter_df = emission_filter_df.reset_index()

emission_filter_df1 = emission_filter_df.rename(columns={'index': 'id'})

emission_filter_df1 = emission_filter_df1 .set_index('id')
```

```python
emission_filter_df1.head()
```

```python
# Merging Dataframes on country and year
merge_df = pd.merge(emission_filter_df1, final_temp_df_new1, how='left', on=['country', 'year'])
merge_df = merge_df[['country', 'co2_emissions', 'temp']]
merge_df
```
